Now we will use summarise and count  the trends we see in the cases which were opened , reopened and disposed  over  the years


```python
# Sum up the total counts for relevant columns
total_reported = df['Cases Reported during the year - Col. (4)'].sum()
total_reopened = df['Cases Reopened for Investigation - Col. (5)'].sum()
total_disposed = df['Total Cases Disposed Off by Police (Col.7+8+15+18) - Col. (19)'].sum()

# Create a DataFrame for plotting
summary_data = pd.DataFrame({
    'Category': ['Total Reported', 'Total Reopened', 'Total Disposed'],
    'Count': [total_reported, total_reopened, total_disposed]
})

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=summary_data, x='Category', y='Count')
plt.title('Overall Numbers of Reported, Reopened, and Disposed Cases')
plt.xlabel('Category')
plt.ylabel('Number of Cases')
plt.show()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(summary_data['Count'], labels=summary_data['Category'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Proportion of Reported, Reopened, and Disposed Cases')
plt.show()
```

Now we check the Highest and Lowest no of cases reported and disposed by the Crime  Heads


```python
# Sum cases by crime head
crime_head_data = df.groupby('Crime Head Col. (2)').sum()

# Sort and get top/bottom 10 crime heads
top_reported = crime_head_data.nlargest(10, 'Cases Reported during the year - Col. (4)')
bottom_reported = crime_head_data.nsmallest(10, 'Cases Reported during the year - Col. (4)')
top_disposed = crime_head_data.nlargest(10, 'Total Cases Disposed Off by Police (Col.7+8+15+18) - Col. (19)')
bottom_disposed = crime_head_data.nsmallest(10, 'Total Cases Disposed Off by Police (Col.7+8+15+18) - Col. (19)')

# Bar plots
plt.figure(figsize=(14, 8))
sns.barplot(data=top_reported.reset_index(), x='Cases Reported during the year - Col. (4)', y='Crime Head Col. (2)')
plt.title('Top 10 Crime Heads with Highest Cases Reported')
plt.xlabel('Number of Cases Reported')
plt.ylabel('Crime Head')
plt.show()

plt.figure(figsize=(14, 8))
sns.barplot(data=bottom_reported.reset_index(), x='Cases Reported during the year - Col. (4)', y='Crime Head Col. (2)')
plt.title('Top 10 Crime Heads with Lowest Cases Reported')
plt.xlabel('Number of Cases Reported')
plt.ylabel('Crime Head')
plt.show()

plt.figure(figsize=(14, 8))
sns.barplot(data=top_disposed.reset_index(), x='Total Cases Disposed Off by Police (Col.7+8+15+18) - Col. (19)', y='Crime Head Col. (2)')
plt.title('Top 10 Crime Heads with Highest Cases Disposed Of')
plt.xlabel('Number of Cases Disposed Of')
plt.ylabel('Crime Head')
plt.show()

plt.figure(figsize=(14, 8))
sns.barplot(data=bottom_disposed.reset_index(), x='Total Cases Disposed Off by Police (Col.7+8+15+18) - Col. (19)', y='Crime Head Col. (2)')
plt.title('Top 10 Crime Heads with Lowest Cases Disposed Of')
plt.xlabel('Number of Cases Disposed Of')
plt.ylabel('Crime Head')
plt.show()

# Pie charts for top and bottom reported cases
plt.figure(figsize=(10, 6))
plt.pie(top_reported['Cases Reported during the year - Col. (4)'], labels=top_reported.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Top 10 Crime Heads with Highest Cases Reported')
plt.show()

plt.figure(figsize=(10, 6))
plt.pie(bottom_reported['Cases Reported during the year - Col. (4)'], labels=bottom_reported.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Top 10 Crime Heads with Lowest Cases Reported')
plt.show()
```

Disposal rates across different crime heads

```python
# Calculate disposal rates
df['Disposal Rate'] = df['Total Cases Disposed Off by Police (Col.7+8+15+18) - Col. (19)'] / df['Total Cases for Investigation (Col.3+Col.4+Col.5) - Col. (6)']

# Sort and get top/bottom 10 disposal rates
top_disposal_rates = df.nlargest(10, 'Disposal Rate')
bottom_disposal_rates = df.nsmallest(10, 'Disposal Rate')

# Bar plots
plt.figure(figsize=(14, 8))
sns.barplot(data=top_disposal_rates, x='Disposal Rate', y='Crime Head Col. (2)')
plt.title('Top 10 Crime Heads with Highest Disposal Rates')
plt.xlabel('Disposal Rate')
plt.ylabel('Crime Head')
plt.show()

plt.figure(figsize=(14, 8))
sns.barplot(data=bottom_disposal_rates, x='Disposal Rate', y='Crime Head Col. (2)')
plt.title('Top 10 Crime Heads with Lowest Disposal Rates')
plt.xlabel('Disposal Rate')
plt.ylabel('Crime Head')
plt.show()

# Box plot to show distribution of disposal rates
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Disposal Rate', y='Crime Head Col. (2)')
plt.title('Distribution of Disposal Rates Across Different Crime Heads')
plt.xlabel('Disposal Rate')
plt.ylabel('Crime Head')
plt.show()
```

Discharging of chargesheeting rates across different crime heads

```python
# Histogram and KDE plot
plt.figure(figsize=(14, 8))
sns.histplot(data=df, x='Chargesheeting Rate (Col.18/19*100) - Col. (23)', kde=True, bins=30)
plt.title('Distribution of Chargesheeting Rates Across Different Crime Heads')
plt.xlabel('Chargesheeting Rate')
plt.ylabel('Frequency')
plt.show()

# Violin plot
plt.figure(figsize=(14, 8))
sns.violinplot(data=df, x='Chargesheeting Rate (Col.18/19*100) - Col. (23)', y='Crime Head Col. (2)')
plt.title('Density and Distribution of Chargesheeting Rates Across Different Crime Heads')
plt.xlabel('Chargesheeting Rate')
plt.ylabel('Crime Head')
plt.show()
```

Pendency Percentage for Different Crime Head

```python

# Calculate pending cases at the end of the year
df['Pending Cases at End of Year'] = df['Cases Pending Invsgn.at end of the year (Col.6-9-19-20) - Col. (22)']

# Bar plot for pending cases
plt.figure(figsize=(14, 8))
sns.barplot(data=df, x='Pending Cases at End of Year', y='Crime Head Col. (2)')
plt.title('Pending Cases at the End of the Year for Different Crime Heads')
plt.xlabel('Number of Pending Cases')
plt.ylabel('Crime Head')
plt.show()

# Distribution plot (histogram and KDE)
plt.figure(figsize=(14, 8))
sns.histplot(data=df, x='Pending Cases at End of Year', kde=True, bins=30)
plt.title('Distribution of Pending Cases at the End of the Year')
plt.xlabel('Number of Pending Cases')
plt.ylabel('Frequency')
plt.show()

# Cumulative distribution function (CDF)
sorted_pending_cases = df['Pending Cases at End of Year'].sort_values()
cdf = sorted_pending_cases.cumsum() / sorted_pending_cases.sum()

plt.figure(figsize=(14, 8))
plt.plot(sorted_pending_cases, cdf, marker='.', linestyle='none')
plt.title('Cumulative Distribution of Pending Cases at the End of the Year')
plt.xlabel('Number of Pending Cases')
plt.ylabel('Cumulative Percentage')
plt.grid(True)
plt.show()
```

Analysis of Charge Sheets Submitted

```python
# Sum cases by crime head
crime_head_data = df.groupby('Crime Head Col. (2)').sum()

# Bar plot for charge sheets submitted from previous year
plt.figure(figsize=(14, 8))
sns.barplot(data=crime_head_data.reset_index(), x='Charge Sheets Submitted - Cases ChargeSheeted Out of cases From Prev. Year - Col. (16)', y='Crime Head Col. (2)')
plt.title('Charge Sheets Submitted from Previous Year by Crime Head')
plt.xlabel('Number of Charge Sheets Submitted')
plt.ylabel('Crime Head')
plt.show()

# Bar plot for charge sheets submitted during the current year
plt.figure(figsize=(14, 8))
sns.barplot(data=crime_head_data.reset_index(), x='Charge Sheets Submitted - Cases Charge sheeted Out of Cases during the year - Col. (17)', y='Crime Head Col. (2)')
plt.title('Charge Sheets Submitted during the Current Year by Crime Head')
plt.xlabel('Number of Charge Sheets Submitted')
plt.ylabel('Crime Head')
plt.show()

# Pie charts for proportions of charge sheets submitted
charge_sheets_data = df[['Charge Sheets Submitted - Cases ChargeSheeted Out of cases From Prev. Year - Col. (16)',
                         'Charge Sheets Submitted - Cases Charge sheeted Out of Cases during the year - Col. (17)']].sum()

plt.figure(figsize=(8, 8))
plt.pie(charge_sheets_data, labels=['Previous Year', 'Current Year'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen'])
plt.title('Proportion of Charge Sheets Submitted from Previous and Current Year')
plt.show()

# Comparison of charge sheets submitted in previous vs. current year
plt.figure(figsize=(14, 8))
charge_sheets_comparison = df[['Crime Head Col. (2)', 'Charge Sheets Submitted - Cases ChargeSheeted Out of cases From Prev. Year - Col. (16)',
                               'Charge Sheets Submitted - Cases Charge sheeted Out of Cases during the year - Col. (17)']]
charge_sheets_comparison = charge_sheets_comparison.melt(id_vars=['Crime Head Col. (2)'], var_name='Year', value_name='Charge Sheets Submitted')

sns.barplot(data=charge_sheets_comparison, x='Charge Sheets Submitted', y='Crime Head Col. (2)', hue='Year')
plt.title('Comparison of Charge Sheets Submitted in Previous vs. Current Year by Crime Head')
plt.xlabel('Number of Charge Sheets Submitted')
plt.ylabel('Crime Head')
plt.show()
```

Final Report Analysis

```python
# Bar plot for final reports
plt.figure(figsize=(14, 8))
final_report_columns = [
    'Final Report - Cases Ended as FR Non Cognizable - Col. (10)',
    'Final Report - Cases Ended as Final Report False - Col. (11)',
    'Final Report - Cases Ended as Mistake of Fact or of law or Civil Dispute - Col. (12)',
    'Final Report - Cases True but Insufficient Evidence or Untraced or No Clue - Col. (13)',
    'Final Report - Cases Abated during Investigation - Col. (14)'
]

final_report_data = df[final_report_columns].sum().reset_index()
final_report_data.columns = ['Final Report Type', 'Count']

sns.barplot(data=final_report_data, x='Count', y='Final Report Type')
plt.title('Different Types of Final Reports Submitted')
plt.xlabel('Number of Final Reports')
plt.ylabel('Final Report Type')
plt.show()

# Pie chart for proportions of final reports
plt.figure(figsize=(10, 6))
plt.pie(final_report_data['Count'], labels=final_report_data['Final Report Type'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Proportion of Different Types of Final Reports')
plt.show()
```
